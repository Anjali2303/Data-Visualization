```python
df.cov()
```

Covariance tells us about how any two features vary or differ from each other

A negative Covariance means that there is an inverse relationship between two features.
A positive Covariance means that there is an direct relationship between two features.
A zero covariance means that there is no relationship between two features.


***fnlwgt and education-num*** have a strong negative covariance whereas ***capital-gain and age*** have a strong positive covariance.
There are no feature pairs with zero covariance.
This means evry feature affects evry other feature in some magnitude.

```python
df.corr()
```

Correlation shows you how the two variables are related or their similarity.

Correlation is expressed on a range from +1 to -1.

Zero indicates a lack of correlation: There is no tendency for the variables to fluctuate in tandem either positively or negatively.

```python
import seaborn as sns

# To alter the figure size
sns.set(rc = {'figure.figsize':(7,5)})

# To change font size and rotations of axis labels
plt.xticks(fontsize = 10, rotation = 90)

# Plotting histplot
sns.histplot(data=df, x='occupation',y='income')
```

Exec-managerial and Prof-Speciality have the highest incomes.

### SCATTER PLOT WITH CUSTOM VARIABLES

```python
import seaborn as sns

sns.set(rc = {'figure.figsize':(20,15)})

x=df['hours-per-week']
y=df['capital-gain']
sizes = 1000

fig, ax = plt.subplots()
ax.scatter(x, y, sizes, c="red", alpha=0.5, marker=r'$\heartsuit$',
           label="Luck")
ax.set_xlabel("occupation")
ax.set_ylabel("captal-gain")
ax.legend()
plt.show()
```

Some more custom symbols are:

'\\alpha', '\\beta', '\gamma', '\sigma', '\infty', '\spadesuit', '\heartsuit', '\diamondsuit', '\clubsuit', '\\bigodot', '\\bigotimes', '\\bigoplus', '\imath', '\\bowtie', '\\bigtriangleup', '\\bigtriangledown', '\oslash', '\ast', '\\times', '\circ', '\\bullet', '\star', '+', '\Theta', '\Xi', '\Phi',  '\$', '\#', '\%', '\S'
